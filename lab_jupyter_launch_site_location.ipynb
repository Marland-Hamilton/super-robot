```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
launch_site_lat = 28.56245 
launch_site_lon = -80.57723

#  closest city coordinates (latitude, longitude)
closest_city_lat = 28.07198
closest_city_lon = -80.61768
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, closest_city_lat, closest_city_lon)

# Create map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

folium.Marker(
    [launch_site_lat, launch_site_lon], 
    popup="Launch Site",
    icon=folium.Icon(color="blue")
).add_to(site_map)

# Add marker for the closest city with distance
city_marker = folium.Marker(
    [closest_city_lat, closest_city_lon],
    popup="Closest City",
    icon=folium.Icon(color="green")
)
city_marker.add_to(site_map)

# Add a marker with the distance to the closest city
city_distance_marker = folium.Marker(
    [closest_city_lat, closest_city_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_to_city),
    )
)
city_distance_marker.add_to(site_map)

# Create a PolyLine (line) between the launch site and the closest city
coordinates = [
    [launch_site_lat, launch_site_lon],  # Launch site coordinates
    [closest_city_lat, closest_city_lon] # Closest city coordinates
]

# Create PolyLine object
lines = folium.PolyLine(locations=coordinates, weight=2, color="blue")

# Add the PolyLine to the site_map
site_map.add_child(lines)

# Display the map (save to HTML)
site_map.save("launch_site_map_with_closest_city_line.html")

site_map
```
