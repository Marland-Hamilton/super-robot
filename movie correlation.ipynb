```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('/Users/marlandhamilton/Downloads/movies.csv')
```

```python
df
```

```python
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing*100))
```

```python
#drop duplicates
df.drop_duplicates()
```

```python
plt.scatter(x=df['budget'], y=df['gross'],)
plt.title('budget vs Gross Earnings')
plt.xlabel('Budget of Film')
plt.ylabel('Gross for Film')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
sns.regplot(data =df, x="score", y="gross",scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#created a datafarme  with only the num ber columns
df2 = df.select_dtypes(include=['number'])
```

```python
df2
```

```python
df2.corr(method ='pearson')
```

```python
df2.corr(method ='kendall')
```

```python
df2.corr(method ='spearman')
```

```python
correlation_matrix = df2.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
#top 15 compaies by gross revenue
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
